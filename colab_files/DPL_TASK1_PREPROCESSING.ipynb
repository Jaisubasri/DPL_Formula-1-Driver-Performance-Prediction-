```python
import os
import pandas as pd
import numpy as np
import glob
from ydata_profiling import ProfileReport
```

#DATA PREPROCESSING

```python
dataset_path='/content/dataset'
data_dir= glob.glob(os.path.join(dataset_path, "*.csv")) #getting all the directory
dataframes = [pd.read_csv(file) for file in data_dir] #contains all the files
```

```python
preprocessed_data_path = '/content/Preprocessed_Dataset'
os.makedirs(preprocessed_data_path, exist_ok=True)


for file in data_dir:

    df = pd.read_csv(file)

    numeric_cols = df.select_dtypes(include=np.number).columns  #for numeric cols
    df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

    categorical_cols = df.select_dtypes(include='object').columns #for categorical cols
    df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

    df.drop_duplicates(inplace=True)  #drop duplicates
    filename = os.path.splitext(os.path.basename(file))[0]

    preprocessed_file_path = os.path.join(preprocessed_data_path, f"{filename}_preprocessed.csv")
    df.to_csv(preprocessed_file_path, index=False)
    print(f"Preprocessed '{filename}' and saved to '{preprocessed_file_path}'")
```

## Generating Profiling reports

```python
preprocessed_data_dir= glob.glob(os.path.join(preprocessed_data_path, "*.csv"))
profiling_reports_dir = '/content/Profiling_Reports'
os.makedirs(profiling_reports_dir, exist_ok=True)
for file in preprocessed_data_dir:
  df = pd.read_csv(file)
  profile = ProfileReport(df, explorative=True)
  filename = os.path.splitext(os.path.basename(file))[0]
  report_path = os.path.join(profiling_reports_dir, f"{filename}_profile.html")
  profile.to_file(report_path)
  print(f"Profiling report for '{filename}' saved to '{report_path}' ✅")
```
